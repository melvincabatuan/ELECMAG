## Maxwell's First Equation

$$ div \, \textbf{D} = \rho $$

where:

$ div $ - divergence operation

$\textbf{D}$ - Electric flux density 

$ \rho $ - volume charge density


````
txt = '''
Maxwell's First Equation states that, the electric flux per unit volume \
leaving a vanishingly small volume unit is exactly equal to \
the volume charge density there.

This equation is called point form of Gauss Law. It applies to electrostatics \
and steady magnetic fields.

Understanding the divergence operation is crucial in the comprehension of \
Maxwell's First Equation.
'''
filename = '3.wav'; 
speak(txt, filename)
Audio(filename, autoplay=False)
````

## Divergence Operation

$$ div \, \textbf{D}_{cartesian} = \dfrac{\partial D_x}{\partial x} + \dfrac{\partial D_y}{\partial y} + \dfrac{\partial D_z}{\partial z} $$

$$ div \, \textbf{D}_{cylindrical} = \dfrac{1}{\rho}\dfrac{\partial}{\partial \rho}(\rho D_{\rho}) + \dfrac{1}{\rho}\dfrac{\partial D_{\phi}}{\partial \phi} + \dfrac{\partial D_z}{\partial z} $$

$$ div \, \textbf{D}_{spherical} = \dfrac{1}{r^2}\dfrac{\partial}{\partial r}(r^2 D_r) + \dfrac{1}{r \sin \theta}\dfrac{\partial }{\partial \theta}(\sin \theta D_{\theta}) + \dfrac{1}{r \sin \theta}\dfrac{\partial D_\phi}{\partial \phi} $$

````
txt = '''
The divergence of the vector flux density is the outflow of flux 
from a small closed surface per unit volume as the volume shrinks to zero.

Note that there is NO direction associated with divergence since it expresses \
the net outflow or inflow.

More intuition can be found in the following Khan Academy series:
'''
filename = '4.wav'; 
speak(txt, filename)
Audio(filename, autoplay=False)
````

### Divergence with Del or Nabla Operator

$$ \nabla = \langle \dfrac{\partial}{\partial x}\mathbf{a_x} + \dfrac{\partial}{\partial y}\mathbf{a_y} + \dfrac{\partial}{\partial z}\mathbf{a_z}  \rangle$$ 

$$ \nabla \cdot \textbf{D} = \dfrac{\partial D_x}{\partial x} + \dfrac{\partial D_y}{\partial y} + \dfrac{\partial D_z}{\partial z} $$

$$ \nabla \cdot \textbf{D} = \dfrac{1}{\rho}\dfrac{\partial}{\partial \rho}(\rho D_{\rho}) + \dfrac{1}{\rho}\dfrac{\partial D_{\phi}}{\partial \phi} + \dfrac{\partial D_z}{\partial z} $$

$$ \nabla \cdot \textbf{D} = \dfrac{1}{r^2}\dfrac{\partial}{\partial r}(r^2 D_r) + \dfrac{1}{r \sin \theta}\dfrac{\partial }{\partial \theta}(\sin \theta D_{\theta}) + \dfrac{1}{r \sin \theta}\dfrac{\partial D_\phi}{\partial \phi} $$

## Example

Find the $\nabla \cdot \mathbf{D}$ at the origin if $\mathbf{D} = e^{-x} \sin y \mathbf{a}_x - e^{-x}
\cos y \mathbf{a}_y + 2z \mathbf{a} z$ C/m$^2$.

````
 def plot_field(u, v, w):
    plt.subplots(figsize=(16,16))
    plt.subplot(221)
    plt.streamplot(x,y,u,v)
    plt.plot(0,0, 'ro')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.subplot(222)
    plt.streamplot(x,y,u,w)
    plt.plot(0,0, 'ro')
    plt.xlabel('x')
    plt.ylabel('z')
    plt.subplot(223)
    plt.streamplot(x,y,v,w)
    plt.plot(0,0, 'ro')
    plt.xlabel('y')
    plt.ylabel('z')
    plt.subplot(224)
    plt.quiver(x,y,u,w)
    plt.plot(0,0, 'ro')
    plt.xlabel('x')
    plt.ylabel('z')
    plt.show()

x,y = np.meshgrid(np.linspace(-10,10,20),np.linspace(-10,10,20))
x,z = np.meshgrid(np.linspace(-10,10,20),np.linspace(-10,10,20))
u = np.exp(-x)*np.sin(y)
v = -np.exp(-x)*np.cos(y)
w = 2*z
plot_field(u, v, w)
````

## Solution

````
from sympy import *
from sympy.vector import CoordSys3D, divergence
init_printing()

def div(D, *vars):
    """Return the scalar divergence of a vector field."""
    return sum( Matrix( [D[i].diff(v) for i,v in enumerate(vars)]) )
````

````
x, y, z = symbols('x y z', real=True)
D = Matrix([exp(-x)*sin(y), -exp(-x)*cos(y), 2*z])
print(f"Answer = {div(D, x, y, z)} C/m^3")
````

````
# Alternative Solution:

R     = CoordSys3D('R') # Cartesian Reference
D     = exp(-R.x)*sin(R.y)*R.i - exp(-R.x)*cos(R.y)*R.j + 2*R.z*R.k 
print(f"Answer = {divergence(D)} C/m^3")
````

````
# Custom Solution

def div_cylindrical(D, *vars):
    '''
    Returns the divergence of a vector field expressed in cylindrical coordinates
    '''
    assert len(vars) == 3
    Cr = (1/vars[0])*(vars[0]*D[0]).diff(vars[0])
    Cp = (1/vars[0])*(D[1]).diff(vars[1])
    Cz = (D[2]).diff(vars[2])
    return (Cr + Cp + Cz)
````

````
r, p, z = symbols('r p z', real=True)

D = Matrix([2*r*z**2*sin(p)**2, r*z**2 *sin(2*p), 2*r**2 *z*sin(p)**2])
divD = div_cylindrical(D, r, p, z)
print(f"Answer = {divD} C/m^3")
````

````
def div_spherical(D, *vars):
    '''
    Returns the divergence of a vector field expressed in spherical coordinates
    '''
    assert len(vars) == 3
    Sr = (1/vars[0]**2)*(vars[0]**2 * D[0]).diff(vars[0])
    St = (1/(vars[0]*sin(vars[1])))*(sin(vars[1])*D[1]).diff(vars[1])
    Sp = (1/(vars[0]*sin(vars[1])))*(D[2]).diff(vars[2])
    return (Sr + St + Sp)
````

````
r, t, p = symbols('r t p', real=True)

D = Matrix([sin(t)*cos(p), cos(t)*cos(p), - sin(p)])
divD = div_spherical(D, r, t, p)
print(f"Answer = {divD} C/m^3")
````

## Example

Evaluate both sides of divergence theorem for
the field $\mathbf{D} = 2xy\mathbf{a}_x + x^2\mathbf{a}_y$ C/m$^2$ and the rectangular parallelepiped formed by the planes $x = 0$ and $1$,$y = 0$ and $2$, and $z = 0$ and $3$. Note: $\mathbf D$ is parallel to the surfaces at $z = 0$ and $z = 3$, thus, $\mathbf{D}.d\mathbf{S}$ is zero there.

````
x,y = np.meshgrid(np.linspace(-10,10,20),np.linspace(-10,10,20))
x,z = np.meshgrid(np.linspace(-10,10,20),np.linspace(-10,10,20))
u = 2*x*y   
v = x*x   
w = 0*x
plot_field(u, v, w)
````

$$ \oint_s \mathbf D \cdot d \mathbf S = \int_v \nabla \cdot \mathbf D \, dv $$

Evaluating the left side:

$$ \oint_s \mathbf D \cdot d \mathbf S = \int_0^3\int_0^2 (D_x)_{x=0} \cdot (-dydz)\mathbf a_x + \int_0^3\int_0^2 (D_x)_{x=1} \cdot (dydz)\mathbf a_x + \int_0^3\int_0^1 (D_y)_{y=0} \cdot (-dxdz)\mathbf a_y  + \int_0^3\int_0^1 (D_y)_{y=2} \cdot (dxdz)\mathbf a_y $$

$$ D_x = 2xy $$

$$ D_y = x^2 $$

## Solution: (Gauss Law Side - Area)

````
x, y, z = symbols('x y z', real=True)

def eval_gauss(D):  # box
    f = D[0].subs(x, 0)
    rear = integrate(-f,(y, 0, 2),(z, 0, 3))
    f = D[0].subs(x, 1)
    front = integrate(f,(y, 0, 2),(z, 0, 3))
    g = D[1].subs(y, 0)
    left = integrate(-g,(x, 0, 1),(z, 0, 3))
    g = D[1].subs(y, 2)
    right = integrate(g,(x, 0, 1),(z, 0, 3))
    return (rear + front + left + right)

D = Matrix([2*x*y, x*x])
eval_gauss(D)
````

## Custom Divergence Functions Summary

```python
from sympy import *
from sympy.vector import CoordSys3D, divergence
init_printing()

def div_rect(D, *vars):
    """Return the scalar divergence of a vector field."""
    return sum( Matrix( [D[i].diff(v) for i,v in enumerate(vars)]) )


def div_cylindrical(D, *vars):
    '''
    Returns the scalar divergence of a vector field expressed in cylindrical coordinates
    '''
    Cr = (1/vars[0])*(vars[0]*D[0]).diff(vars[0])
    Cp = (1/vars[0])*(D[1]).diff(vars[1])
    return (Cr + Cp)

def div_spherical(D, *vars):
    '''
    Returns the scalar divergence of a vector field expressed in spherical coordinates
    '''
    assert len(vars) == 3
    Sr = (1/vars[0]**2)*(vars[0]**2 * D[0]).diff(vars[0])
    St = (1/(vars[0]*sin(vars[1])))*(sin(vars[1])*D[1]).diff(vars[1])
    Sp = (1/(vars[0]*sin(vars[1])))*(D[2]).diff(vars[2])
    return (Sr + St + Sp)
```
